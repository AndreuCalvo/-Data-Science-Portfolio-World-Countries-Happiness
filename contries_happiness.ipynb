The results of this data missing imputation are not good. We want to respect the tendecy among the regions, and using KKNImputer doesn't happen. We try another approach, we will use a mean imputation segmented per region

```python
#implement mean imputation to family segmented per region
data_year['Family'] = data_year.groupby('Region')['Family'].transform(lambda x: x.fillna(x.mean()))
plt.figure(figsize=(10, 5))
sns.boxplot(x='Year', y='Family', data=data_year, hue='Region', palette='viridis')
plt.title('Family per Year and Region')
plt.show()
```

# Feature Selection

```python
# Select features with high correlation to happiness score 
threshold = 0.5 
high_corr_features = matrix_corr.index[abs(matrix_corr['happy']) > threshold].tolist() 

# Remove the target variable from the list 
high_corr_features.remove('happy') 
# Check for multicollinearity 
for feature in high_corr_features: 
    if any(abs(matrix_corr[feature][high_corr_features]) > 0.8):
        high_corr_features.remove(feature)
        
print("Selected features:", high_corr_features)
```

```python
X = data_happiest_saddest_countries[high_corr_features]
y = data_happiest_saddest_countries['happy']
```

# Models - Logistic Regression

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)
```

```python
model_log_reg = LogisticRegression()
model_log_reg.fit(X_train, y_train)
y_pred_log_reg = model_log_reg.predict(X_test)
```

```python
# Check the accuracy of the model
accuracy = accuracy_score(y_test, y_pred_log_reg)
print("Accuracy:", accuracy)

coefficients = pd.DataFrame({ 'Feature': X.columns, 'Coefficient': model_log_reg.coef_.ravel() })
print(coefficients)
```

Now we want to apply this logistic regression to the entire dataset.

```python
happy_pred_log_reg = model_log_reg.predict(data_country_imputed[high_corr_features])
happy_pred_log_reg
```

```python
data_country_imputed_lr = data_country_imputed.copy()
data_country_imputed_lr['happy'] = happy_pred_log_reg
```

```python
# show me the relationship between the region and the happiness
plt.figure(figsize=(10, 5))
sns.countplot(x='region', data=data_country_imputed_lr, hue='happy', palette='viridis')
plt.title('Happiness per Region')
plt.xticks(rotation=90)
plt.show()
```

# Metrics - Logistic Regression

Now we compare the results of this model using the data_year dataset. First of all we select a treeshold of the meaning of the happiness score (5.379018) to decide if a country is happy or not.

```python
data_year['happy_real'] = data_year['Happiness Score'].apply(lambda x: 0 if x < 5.37 else 1)
data_year_2019 = data_year[data_year['Year'] == 2019]
data_year_2019['happy_real'].value_counts()
data_year_y_real = data_year_2019[['Country', 'happy_real']]
data_year_y_real
```
